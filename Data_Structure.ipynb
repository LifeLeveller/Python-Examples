#Linked List

```python
#LINKED LIST
#Create an empty linked list
l1=[]
#add some string elements to the list
l1.append("America")
l1.append("Japan")
l1.append("India")
#Display the list
print("The existing list is ",l1)

#Display menu
choice=0
while choice<5:
	print("Linked List operations")
	print("1 Add Element")
	print("2 Remove Element")
	print("3 Replace Element")
	print("4 Search for an Element")
	print("5 Exit")
	choice=int(input("Enter your choice: "))
	#Perform operations based on choice
	if choice == 1:
		#Add Element
		elem = input("Enter the country name: ")
		try:
			pos = int(input("Enter the position: "))
			l1.insert(pos,elem)
		except ValueError:
			print("Invalid position")
		else:
			print("The list is ",l1)
	elif choice==2:
		#Remove an element
		try:
			elem = input("Enter the country name: ")
			l1.remove(elem)
		except ValueError:
			print("Country name not Found")
		else:
			print("The list is ",l1)
	elif choice==3:
		#Replace element
		try:
			elem = input("Enter the country name: ")
			pos = int(input("Enter the position: "))
			l1.remove(elem)
			l1.insert(pos,elem)
		except ValueError:
			print("Country name not Found")
		else:
			print("The list is ",l1)
	elif choice==4:
	#search for the index of the element
		elem = input("Enter the country name: ")
		try:
			pos=l1.index(elem)
			print("The position of the country name is : ",pos)
		except ValueError:
			print('Country name not found')
	else:
		break

print("The final list is ",l1)
```

#stacks

```python
#Stacks
class stack:
	def __init__(self):
		self.st=[]
	def isempty(self):
		return self.st == []
		
	def push(self,element):
		#Insert the element at the top of the stack
		self.st.append(element)
		
	def pop(self):
		#Removing the topmost element of the stack
		if self.isempty():
			return -1
		else:
			return self.st.pop()
			
	def peep(self):
		#Returning the topmost element without deleting from the stack
		#n-1th element when there are n elements
		if self.isempty():
			return -1
		else:
			n=len(self.st)
			return self.st[n-1]
		
	def search(self,element):
		#Knowing the postion of the element in the stack from the top of the stack
		if self.isempty():
			return -1
		else:
			try:
				#index return the postion number 'n' of the element from the beginning 
				#(or Bottom)of the stack
				n=self.st.index(element)
				return len(self.st)-n
				#position from the top of the stack is:size of stack-n
			except ValueError:
				return -2
				
	def display(self):
		return self.st
#Importing the created class from the file
#from cl_obj1 import stack

s = stack()

#Display menu
choice=0
while choice<5:
	print("Stack operations")
	print("1 Push Element")
	print("2 Pop Element")
	print("3 Peep Element")
	print("4 Search for an Element")
	print("5 Exit")
	choice=int(input("Enter your choice: "))
	#Perform operations based on choice
	if choice == 1:
		#Add Element
		elem = int(input("Enter the element "))
		s.push(elem)
		print(s.display())
		
	elif choice==2:
		#Remove an element
		elem = s.pop()
		if elem == -1:
			print("The stack is empty")
		else:
			print("The popped element = ",elem)
	elif choice==3:
		elem = s.peep()
		if elem == -1:
			print("Stack is empty")
		else:
			print("The topmost element= ",elem)
	elif choice==4:
	#search for the index of the element
		elem = input("Enter the element: ")
		pos = s.search(elem)
		if pos == -1:
			print("The stack is empty")
		elif pos ==-2:
			print("Element is not found in the stack")
		else:
			print("The element {} is found at position {}".format(elem,pos))
	else:
		break

print("The final stack is ",s.display())
		
```

#Queue - FIFO

```python
#Queue - First In First Out
class queue:
	def __init__(self):
		self.qu =[]
	def is_empty(self):
		return self.qu == []
	def add(self,element):
		self.qu.append(element)
	def delete(self):
		if self.is_empty():
			return -1
		else:
			return self.qu.pop(0)
	def search(self,element):
		if self.is_empty():
			return -1
		else:
			try:
				n=self.qu.index(element)
				return n+1
			except ValueError:
				return -2
	def display(self):
		return self.qu

q=queue()
#Display menu
choice=0
while choice<4:
	print("Queue operations")
	print("1 Add Element")
	print("2 Delete Element")
	print("3 Search for Element")
	print("4 Exit")
	choice=int(input("Enter your choice: "))
	#Perform operations based on choice
	if choice == 1:
		#Add Element
		elem = int(input("Enter the element "))
		q.add(elem)
		print(q.display())
		
	elif choice==2:
		#Remove an element
		elem = q.delete()
		if elem == -1:
			print("The queue is empty")
		else:
			print("The removed element = ",elem)
	elif choice==3:
	#search for the index of the element
		elem = int(input("Enter the element: "))
		pos = q.search(elem)
		if pos == -1:
			print("The queue is empty")
		elif pos ==-2:
			print("Element is not found in the queue")
		else:
			print("The element {} is found at position {}".format(elem,pos))
	else:
		break

print("The final queue is ",q.display())
		
```
